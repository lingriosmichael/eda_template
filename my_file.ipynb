# Column Names and descriptions for King County Data Set
- **id** - unique identifier for a house

- **pricePrice** - prediction target
- **bedroomsNumber** - # of bedrooms
- **bathroomsNumber** - # of bathrooms
- **sqft_livingsquare** - footage of the home
- **sqft_lotsquare** - footage of the lot
- **floorsTotal** - floors (levels) in house
- **waterfront** - House which has a view to a waterfront
- **view** - quality of view
- **condition** - How good the condition is ( Overall )
- **grade** - overall grade given to the housing unit, based on King County grading system
- **sqft_above** - square footage of house apart from basement
- **sqft_basement** - square footage of the basement
- **yr_built** - Built Year
- **yr_renovated** - Year when house was renovated
- **zipcode** - zip
- **lat** - Latitude coordinate
- **long** - Longitude coordinate
- **sqft_living15** - The square footage of interior housing living space for the nearest 15 neighbors
- **sqft_lot15** - The square footage of the land lots of the nearest 15 neighbors

```python
# dropping unwanted columns for further analysis (id, id.1, sqft_above, sqft_basement, sqft_living15, sqft_lot15)

columns_needed = [
    'house_id',
    'date',
    'price',
    'zipcode',
    'lat',
    'long',
    'sqft_living',
    'sqft_lot',
    'bedrooms',
    'bathrooms',
    'floors',
    'grade',
    'condition',
    'waterfront',
    'view'
]

df_copy = df_copy[columns_needed]
df_copy
```

```python
# in precipitation column, replace NaN with 0, in waterfront column
## df_copy.fillna({"waterfront":0})

df_copy["view"] = df_copy["view"].fillna(0)
df_copy["waterfront"] = df_copy["waterfront"].fillna(0)
```

```python
# Convert specific columns to int64
df_copy = df_copy.astype({
    'price': 'int64',
    'house_id': 'int64',
    'sqft_living': 'int64',
    'sqft_lot': 'int64',
    'floors': 'int64',
    'waterfront': 'int64',
    'view': 'int64',
    'condition': 'int64',
    'grade': 'int64'
})


```

```python
## Date column needs to be changed to date type.


## Keep the last record for each sale per house.

df_copy = (
    df_copy.sort_values("date")
      .drop_duplicates(subset="house_id", keep="last")
)
```
