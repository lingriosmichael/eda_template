```python
# Step1> Define the Luxury Price Threshold

# Load your data
df_luxury_price = df_copy.copy()
min_price = df_luxury_price['price'].min()
max_price = df_luxury_price['price'].max()
mean_price = df_luxury_price['price'].mean()
print(f"Minimum house price: ${min_price:,.2f}")
print(f"Maximum house price: ${max_price:,.2f}")
print(f"Mean house price: ${mean_price:,.2f}")  

# Calculate the 90th percentile
luxury_threshold = np.percentile(df_luxury_price['price'], 90)
print(f"Houses priced above ${luxury_threshold:,.2f} are considered luxury.")
```

Hier ist die schrittweise Aufschlüsselung:

1. Daten kopieren
df_luxury_price = df_copy.copy()
Zuerst wird eine Kopie des bestehenden DataFrames (df_copy) erstellt. Das ist eine „best practice“, um sicherzustellen, dass die ursprünglichen Daten nicht versehentlich verändert werden, während du mit der neuen Variable arbeitest.

2. Den Schwellenwert berechnen
luxury_threshold = np.percentile(df_luxury_price['price'], 90)
Hier passiert die eigentliche Magie mit der Bibliothek NumPy (np):
Es wird das 90. Perzentil der Spalte price berechnet.
Das bedeutet: Der Code sucht den Preiswert, der höher ist als 90 % aller anderen Hauspreise im Datensatz. Nur die teuersten 10 % der Häuser liegen über diesem Wert.

3. Ergebnis ausgeben
print(f"Houses priced above ${luxury_threshold:,.2f} are considered luxury.")
Dieser Teil gibt das Ergebnis in einem lesbaren Format aus.

Das :,.2f sorgt dafür, dass der Preis mit Tausendertrennzeichen und zwei Nachkommastellen angezeigt wird (z. B. $1,250,000.00).

```python
# Step 2: Filter for Luxury Houses
# Create a new DataFrame containing only the houses that meet or exceed this price threshold.

# Create the luxury dataframe
luxury_df = df_luxury_price[df_luxury_price['price'] >= luxury_threshold]

# How many luxury houses are there?
num_luxury = luxury_df.shape[0]
print(f"There are {num_luxury} luxury houses.")
# Using .shape[0] gives you the number of rows, 
# which corresponds to the number of individual houses in that category.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# 1. Get the counts (value_counts sorts from highest to lowest by default)
location_dist = luxury_df['zipcode'].value_counts()

plt.figure(figsize=(14, 7))

# 2. Use the 'order' parameter to explicitly set the sequence from highest to lowest
sns.barplot(
    x=location_dist.index, 
    y=location_dist.values, 
    palette='magma',
    order=location_dist.index  # This ensures the bars follow the sorted order of the counts
)

plt.title('Luxury House Distribution (Highest to Lowest Concentration)', fontsize=16)
plt.xlabel('Zipcode', fontsize=12)
plt.ylabel('Number of Houses', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('sorted_zipcode_distribution.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Set a clean aesthetic
sns.set_theme(style="whitegrid")

# Data source
prices = df_luxury_price['price']

# Create a figure with two vertical subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# --- Subplot 1: Entry Level Luxury (0 to 889,000) ---
sns.histplot(
    prices, 
    bins=20, 
    binrange=(0, 889000), 
    ax=ax1, 
    color='skyblue', 
    edgecolor='black'
)
ax1.set_title('Distribution: Entry-Level Luxury (0 - 889k)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Price ($)')
ax1.set_ylabel('Number of Houses')
ax1.set_xlim(0, 889000)

# --- Subplot 2: High-End Luxury (889,000 to 8,000,000) ---
sns.histplot(
    prices, 
    bins=40, 
    binrange=(889000, 8000000), 
    ax=ax2, 
    color='royalblue', 
    edgecolor='black'
)

# Set title and labels
ax2.set_title('Distribution: High-End Luxury (889k - 8M)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Price (Millions of USD)', fontsize=12)
ax2.set_ylabel('Number of Houses')
ax2.set_xlim(889000, 8000000)

# Format the x-axis to show Millions (e.g., 2.0M instead of 2,000,000)
ax2.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'${x/1_000_000:.1f}M'))

# Clean up layout
sns.despine()
plt.tight_layout()
plt.show()
```
